```python
# Importing the necessary libraries
# What's tensorflow, keras and Numpy? Let's go to the slides

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers import Dense
import numpy as np

# Say we wish to find out the relationship, y = mx + c, between our inputs, xs, and outputs, ys
# So that we can predict ys from xs
# This is quintessentially what ML is!
# But of course, this is a very simple example
# Since deep learning is a type of ML, we can use neural networks to help us estimate m and c

# here's the input data
# Can we guess what the actual relationship is? Let's go to the slides
xs = np.array([-2.0, -1.0, 0.0, 1.0, 2.0, 3.0], dtype=float)
ys = np.array([-8.0, -5.0, -2.0, 1.0, 4.0, 7.0], dtype=float)

# We build our model
# Let's go through this line by line

# Our inputs are batches of a single value each, so we specify as a tuple (1,) using the Input() function
inputs = keras.Input(shape = (1,))

# We define our layers using the keras.layers API
# We specify using a Dense layer. Dense means a set of fully connected neurons, i.e. 
# where each neuron is connected to every neuron in the previous layer
# Our Dense layer here has units = 1 specified, which means we just have the one dense 
# layer and one neuron in our entire neural network (see slide)
outputs = layers.Dense(1)(inputs)

# create our model from inputs and outputs, using the Model class
model = keras.Model(inputs = inputs, outputs = outputs)

# configures the model for training, using compile() method
model.compile(optimizer='sgd', loss='mean_absolute_error')
# What do the optimizer and loss mean? Let's go to our slides

# start the learning with the fit() method, over 500 iterations in this case
# we save the results into "history" so we can plot later 
history = model.fit(xs, ys, epochs=500)

# One way to interpret what we are doing here is: fit the values of X to the corresponding values of Y
# Try this fitting 500 times 
# So on the first try, we will start with randomised weights, e.g. Y = 5X + 5
# We will measure the loss, feed the result to the optimizer, and generate another guess
# This process is then repeated, and the loss should go down with more and more repeats (epochs)
```

# Visualising our results

```python
# Let us see the trend of how loss is changing with epochs

# Import the matplotlib library for plotting
import matplotlib.pyplot as plt

# create a plotting function
# the function will take in three arguments: (1) process, which is storing the results of our training
# (2) metric, which is the performance metric that we want to plot against epoch #
# (3) filename, which is the name of the file where we save the results
def plot_graphs(process, metric, filename):      
    plt.plot(process.history[metric])
    plt.title(filename.split('.')[0])
    plt.xlabel("Epochs")
    plt.ylabel(metric)
    plt.savefig(filename)
    plt.show()
```

```python
# make use of the function to plot how loss is changing with epochs
plot_graphs(history, 'loss', 'First NN - Loss.png')

# Look at the trend below. Is it what we expect? 
# We have also saved a copy of our figure. Go to RHS -> Output -> Find our filename -> Ellipsis -> Download
```

```python
# we will need a callback function (we will discuss more on this in Class 2/3)
# But essentially a callback is a function that can be called repeatedly during the training at key points
# Specifically, we need a custom callback that can be created with a LambdaCallback object

# do the necessary imports, including LambdaCallback
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers import Dense
import numpy as np
from keras.callbacks import LambdaCallback

# create an empty dictionary to store our weights at each epoch 
weights_dict = {}

# define our custom callback 
# at the end of every epoch, we will update our dictionary with the weights of the model 
weight_callback = LambdaCallback(on_epoch_end=lambda epoch, logs: weights_dict.update({epoch:model.get_weights()}))

# specify our training data
xs = np.array([-2.0, -1.0, 0.0, 1.0, 2.0, 3.0], dtype=float)
ys = np.array([-8.0, -5.0, -2.0, 1.0, 4.0, 7.0], dtype=float)

# We build our model, as before
inputs = keras.Input(shape = (1,))
outputs = layers.Dense(1)(inputs)

model = keras.Model(inputs = inputs, outputs = outputs)
model.compile(optimizer='sgd', loss='mean_absolute_error')

# specify the callback in fit()
model.fit(xs, ys, epochs=500, callbacks=weight_callback)

# we print out the weights stored in our dictionary
for epoch, weights in weights_dict.items():
    print("Value of m for epoch #",epoch+1)
    print(weights[0])
    print("Value of c for epoch #",epoch+1)
    print(weights[1])
    
# run the code chunk
```
