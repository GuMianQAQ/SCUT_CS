### Vectorization : encoding language into numbers

```python
# import tensorflow and keras
import tensorflow as tf
from tensorflow import keras

# import the Tokenizer class from the keras preprocessing library for word tokenization
from tensorflow.keras.preprocessing.text import Tokenizer

# we have two sample sentences, stored in the variable "sentences"
sentences = [
'I am very happy today',
'I am very sad today'
]

# create a Tokenizer object (instantiating a class gives an object) 
# and specify the number of words it can tokenize
# note that setting num_words = 100 means that the most common 99 words will be kept, i.e. num_words - 1
# we have a very small corpus in our sample sentences, so no issue with the limit here
tokenizer = Tokenizer(num_words = 100)

# we use the fit_on_texts() method for our tokenizer object to create the tokenized word index
tokenizer.fit_on_texts(sentences)

# save our word index, by accessing the word_index attribute of our tokenizer object 
word_index = tokenizer.word_index
# let's take a look at our word index
print(word_index)
```

```python
# if we expand our corpus with another sentence 
sentences = [
'I am very happy today',
"I am very sad today, aren't I?",
"Am I happy today?"
]

# create our tokenizer object
tokenizer = Tokenizer(num_words = 100)
# generate the word index
tokenizer.fit_on_texts(sentences)
# retrieve our word index for display
word_index = tokenizer.word_index
print(word_index)

# Qn 1: What do we notice? Let's go to our slides
```

### Write out the three sentences using the word index!

```python
# the text_to_sequences method allows us to encode our sentences into sequences of numbers, based on our word index
sequences = tokenizer.texts_to_sequences(sentences)
print(sequences)

# Is the result the same as what we jotted down by hand?
# Let's go straight to the next code cell
```

### Let's apply our word index to tokenize other sentences

```python
# Let's say we have a couple of test sentences, as follows
test_data = [
'Today I am melancholic',
'Will I be happy tomorrow?'
]

# generate the vectorized sequences
test_sequences = tokenizer.texts_to_sequences(test_data)
print(word_index)
print(test_sequences)

# Qn 2: What do we notice here? Swap the indices for words; do the sentences make sense? <slides>
```

```python
# here, we specify an OOV token
# note that the OOV token counts towards the 99 word limit that will be retained, i.e. one of your words is <OOV>
# when we instantiate our tokenizer object, we specify the OOV token as well 
tokenizer = Tokenizer(num_words = 100, oov_token="<OOV>")

# generate our word index
tokenizer.fit_on_texts(sentences)
word_index = tokenizer.word_index


# we apply our word index again on our test sentences, to vectorize them 
test_sequences = tokenizer.texts_to_sequences(test_data)
print(word_index)
print(test_sequences)
# notice that by default, the index 1 is reserved for the OOV token

# Qn 3: Convert the vectors back into sentences. What do we have? <slides>
```

# Padding

```python
# we add a fourth, longer sentence to illustrate padding
sentences = [
'I am happy today',
'I am sad today',
'Am I happy today?',
'I feel happy taking a walk outside today'
]

# create our Tokenizer object
tokenizer = Tokenizer(num_words = 100, oov_token="<OOV>")
# generate our word index
tokenizer.fit_on_texts(sentences)
word_index = tokenizer.word_index
# vectorize our sentences 
sequences = tokenizer.texts_to_sequences(sentences)

print(sequences)
# we see that the sequences are of varying lengths; we need them to be the same length for
# training neural networks
```

```python
# to make these sentences the same length, we can make use of the pad_sequences function
# let's import it first
from tensorflow.keras.preprocessing.sequence import pad_sequences

# we can then simply call the function to convert our sequences into a padded set
padded = pad_sequences(sequences)
print(padded)
# we see a nicely-formatted set of sequences below
```

#### Let's go back to our slides to discuss the results

```python
# To change the default to post padding, we can simply do the following: 
padded = pad_sequences(sequences, padding = 'post')
print(padded)
```

```python
# What if you have a sentence that is way longer than the rest
# We will end up with a lot of padding 
# To address this concern, we can specify a maximum length for sequences when
# calling pad_sequences, as shown below: 

# preventing the longest sentence from causing too much padding, i.e. long sequences that are 
# computationally intensive to process
padded = pad_sequences(sequences, padding='post', maxlen=5)
print(padded)

# we see that the padded sequences are all the same length, and there isn't too much padding either
```

```python
# Qn 4: What do we notice about the truncating to the max length? Why might this be undesirable? <slides>
```

```python
# truncating the longest sentence from the end, instead of beginning
padded = pad_sequences(sequences, padding='post', maxlen=6, truncating='post')
print(padded)
```
